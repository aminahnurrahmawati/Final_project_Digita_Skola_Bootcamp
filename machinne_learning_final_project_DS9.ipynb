Dealing with categorical data

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
#transforming categorical train data to numerical
train['NAME_CONTRACT_TYPE'] = labelencoder.fit_transform(train['NAME_CONTRACT_TYPE'])
train['CODE_GENDER'] = labelencoder.fit_transform(train['CODE_GENDER'])
train['FLAG_OWN_CAR'] = labelencoder.fit_transform(train['FLAG_OWN_CAR'])
train['FLAG_OWN_REALTY'] = labelencoder.fit_transform(train['FLAG_OWN_REALTY'])
train['NAME_TYPE_SUITE'] = labelencoder.fit_transform(train['NAME_TYPE_SUITE'])
train['NAME_INCOME_TYPE'] = labelencoder.fit_transform(train['NAME_INCOME_TYPE'])
train['NAME_EDUCATION_TYPE'] = labelencoder.fit_transform(train['NAME_EDUCATION_TYPE'])
train['NAME_FAMILY_STATUS'] = labelencoder.fit_transform(train['NAME_FAMILY_STATUS'])
train['NAME_HOUSING_TYPE'] = labelencoder.fit_transform(train['NAME_HOUSING_TYPE'])
train['OCCUPATION_TYPE'] = labelencoder.fit_transform(train['OCCUPATION_TYPE'])
train['WEEKDAY_APPR_PROCESS_START'] = labelencoder.fit_transform(train['WEEKDAY_APPR_PROCESS_START'])
train['ORGANIZATION_TYPE'] = labelencoder.fit_transform(train['ORGANIZATION_TYPE'])
train['HOUSETYPE_MODE'] = labelencoder.fit_transform(train['HOUSETYPE_MODE'])
train['WALLSMATERIAL_MODE'] = labelencoder.fit_transform(train['WALLSMATERIAL_MODE'])
train['EMERGENCYSTATE_MODE'] = labelencoder.fit_transform(train['EMERGENCYSTATE_MODE'])
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
test_trans = sc.fit_transform(test)
```

Replace Missing Values with median

```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(missing_values= np.nan, strategy='median')
```

```python
x_train = imputer.fit_transform(x_train)
```

```python
test_trans= imputer.fit_transform(test_trans)
```

# Applying ML model 

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver ='liblinear', random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(test_trans)
```

```python
y_pred
```

```python
Target_prediksi = pd.DataFrame({"TARGET" : y_pred})
```

```python
Target_prediksi
```

```python
prediksi_and_testing_data = [Target_prediksi, test]
prediksi_and_testing_data = pd.concat(prediksi_and_testing_data).drop_duplicates().reset_index(drop=True)
```

Sending Result to Postgre

```python
from sqlalchemy import create_engine
#dialect+driver://username:password@host:port/database

engine=create_engine("postgresql+psycopg2://postgres:mekanika@localhost:5432/final_project")
```

```python
try:
    prediksi_and_testing_data.to_sql('hasil_prediksi_ML', engine, if_exists= 'replace', index= False) #nama tablenya HomeCredit_columns_description

except:
    print("Sorry, some error has occurred!")

finally:
    engine.dispose()
```
